```python
import csv
from selenium import webdriver

driver=webdriver.Chrome(executable_path=r"E:\data prepration\driver_chrome\New folder\chromedriver.exe")



driver.get("https://www.amazon.in/gp/bestsellers/books/")


products = driver.find_elements_by_xpath("//div[@class='a-cardui _cDEzb_grid-cell_1uMOS p13n-grid-content']")


names=[]
prices=[]


counter=0
for i in products:
    try:
        my_name = i.find_element_by_xpath(".//div[@class='_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y']")
        my_price = i.find_element_by_xpath(".//span[@class='a-size-base a-color-price']")
        print(f"number of book{counter}")
        print("-----------")
        print (my_name.text)
        print (my_price.text)
        print("-------------------------------------------------")
        names.append(my_name.text)
        prices.append(my_price.text)
        counter=counter+1
    except:
        print("-------------------------------------------------")
        print("-------------------------------------------------")
        print('check')
        print("-------------------------------------------------")
        print("-------------------------------------------------")
        print("-------------------------------------------------")
        
        
driver.close()        
```

```python
with open(r'E:\data prepration\data_from_scraping\book.csv', 'w',encoding='utf-8',newline='') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')
    csvwriter.writerow(['product','Price'])
    for i in range(1,len(names)):
        csvwriter.writerow([names[i],prices[i]])
```
